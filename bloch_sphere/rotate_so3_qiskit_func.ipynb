Rotating around the Bloch sphere

```python
import numpy as np
from qiskit import QuantumCircuit, QuantumRegister, execute
from qiskit.visualization import plot_bloch_multivector

qr = QuantumRegister(1)
circ = QuantumCircuit(qr)
```

```python

circ.x(0)

# Rotation in y
circ.ry(1,qr[0])

# Rotation in x
circ.rx(1,qr[0])

# Rotation in z
circ.rz(1,qr[0])

# Rotation in xyz
circ.rv(1,1,1,qr[0])

# Rotation with 2 Euler angles and state
# angles named: theta, phi, qubit
circ.r(1,1,qr[0])


# Rotation with 3 Euler angles and state
# angles named: theta, phi, lam, qubit
circ.u(3,2,1,qr[0])


# Draw the circuit
circ.draw
plot_bloch_multivector(circ)
```

```python
# Use the BasicAer statevector_simulator backend
from qiskit import BasicAer
backend_sv_sim = BasicAer.get_backend('statevector_simulator')

job_sim = execute(circ, backend_sv_sim)
result_sim = job_sim.result()
quantum_state = result_sim.get_statevector(circ, decimals=3)

# Output the quantum state vector
quantum_state
```

```python
# Plot the state vector on a Bloch sphere
from qiskit.tools.visualization import plot_bloch_multivector
plot_bloch_multivector(quantum_state)
```
